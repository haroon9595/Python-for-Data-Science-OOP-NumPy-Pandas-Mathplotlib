```python
months = ["jan","feb","march"]
sales = [10000,15000,12000]
plt.plot(months,sales,color= "blue",marker = "o", linestyle="--" , linewidth = 2  )
plt.xlabel("month")
plt.ylabel("sales per month")
plt.title("Monthly show data")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["jan","feb","march"]
sales = [10000,15000,12000]

plt.plot(months, sales,
         color="blue",
         marker="o",
         linestyle="--",
         linewidth=2,
         label="Sales Data")   # ← label add kiya

plt.xlabel("month")
plt.ylabel("sales per month")
plt.title("Monthly show data")
plt.legend(loc="upper left",fontsize=10)
plt.grid(linewidth=1,color="gray",linestyle=":")
#plt.xlim(0,4)
#plt.ylim(1,2000)
# change x axis names with new names
plt.xticks(["jan","feb","march"],["M1","M2","M3"])


plt.show()
```

```python
months = ["Jan", "Feb", "Mar"]

sales = [10000, 15000, 12000]
profit = [4000, 7000, 6000]
expenses = [6000, 8000, 7500]
plt.grid(color="gray" , linestyle=":" , linewidth = 1)

plt.plot(months, sales, label="Sales" , marker = "o")
plt.plot(months, profit, label="Profit" ,marker= "<")
plt.plot(months, expenses, label="Expenses" , marker = '.')

plt.legend()
plt.xlim(0,3)
plt.t
plt.show()
```

## pie chart


```python
regions = ["Pakistan","India","Usa","China"]
sales = [1000,2000,3000,4000]
plt.pie(sales , labels =  regions, autopct = "%1.1f%%" , colors=['green','pink','orange','red'] )
plt.legend()
plt.show()
```

```python
regions = ["Pakistan","India","Usa","China"]
sales = [1000,2000,3000,4000]
plt.pie(sales , labels =  regions, autopct = "%1.2f%%" , startangle=90)
plt.legend()
plt.show()
```

## histrogram

```python
marks = [20,21,22,23,30,31,33,40,45,47,48,50,52,55,57]
plt.hist(marks, bins=5, edgecolor="black", color="orange")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.title("Marks Distribution")
plt.show()
```
