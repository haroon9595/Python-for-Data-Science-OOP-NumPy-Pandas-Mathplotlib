# Body mass index (BMI) Calculator

```python
weight = float(input("Enter your weight in kg: "))

print("Select height unit:")
print("1. Feet & Inches")
print("2. Meters")
choice = int(input("Enter choice (1 or 2): "))

if choice == 1:
    feet = int(input("Enter your height (feet): "))
    inches = int(input("Enter remaining inches: "))
    total_feet = feet + inches / 12
    meters = total_feet * 0.3048   # 1 foot = 0.3048 meters
elif choice == 2:
    meters = float(input("Enter your height in meters: "))
else:
    print("❌ kaka sahi number daal")
    exit()

# BMI calculation
bmi = weight / (meters ** 2)

print(f"\nYour BMI is: {bmi:.2f}")

# BMI Categories
if bmi < 18.5:
    print("Underweight – khao piyo sehat banao 🍗🥛")
elif bmi < 25:
    print("Normal weight ✅")
elif bmi < 30:
    print("Overweight ⚠️")
else:
    print("Obese – mota sand 🐂")
```

# intrest rate calculator
## simple intrest 
##### S.I = PRINCIPAL AMOUNT x RATE OF INTREST X TIME PERIOD / 100


```python
intrest_rate = 8
principal_amount = 1500000
time_period = 3
threshold_amount = 120000

simple_intrest1= (intrest_rate*principal_amount*time_period) / 100
simple_intrest= (intrest_rate*principal_amount*time_period) // 100

print(f"simple intreset rate is {simple_intrest1} % in floating")
print(f"simple intreset rate is {simple_intrest} % in integers")


if simple_intrest < threshold_amount:
    print("intrest rate is less than the threshold amount")
elif simple_intrest>threshold_amount:
    print("intrest rate is greater than the threshold the amount")

print(simple_intrest>threshold_amount)
```

```python
def result(marks):
    average = sum(marks)/len(marks)
    highest = max(marks)
    lowest = min(marks)
    fail =0
    Pass = 0
    for i in marks:
        if i < 50:
            fail=fail+1
            
    Pass = len(marks) - fail

    return average , highest , lowest , fail ,Pass 
```

```python
exam_marks = [30,40,50,20,60,90,95,80,70]
average , highest , lowest , fail ,Pass = result(exam_marks)
print("average: ",round(average,2))
print("highest: ",highest)
print("lowest: ",lowest)
print("pass: ",Pass)
print("fail: ",fail)
```
