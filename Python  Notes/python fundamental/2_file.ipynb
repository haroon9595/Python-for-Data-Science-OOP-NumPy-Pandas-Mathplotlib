## Age Group Clasification

```python
num_people = int(input("entet the number of people"))
age_list = []
for i in range(num_people):
    age = int(input(f"enter the age of {i+1} person: "))
    age_list.append(age)
age_group = []
for i in range(num_people):
    if age_list[i] <18:
        age_group.append("child")
    elif 18<=age_list[i]<65:
        age_group.append("Adults")
    else:
        age_group.append("senior citizon")
print(age_list)
print(age_group)        
```

```python
num_people = int(input("entet the number of people"))
age_list = []
for i in range(num_people):
    age = int(input(f"enter the age of {i+1} person: "))
    age_list.append(age)
age_group = []
for age in age_list:
    if age <18:
        age_group.append("child")
    elif 18<=age<65:
        age_group.append("Adults")
    else:
        age_group.append("senior citizon")
adult_count = 0
child_count= 0
senior_count = 0     
for i in age_group:
    if i == "Adults":
        adult_count = adult_count +1 
    if i == "child":
        child_count = child_count +1
    if i == "senior citizon":
        senior_count = senior_count+1       

        
for i in range(num_people):
    print(f"{i+1} | age | {age_list[i]} | {age_group[i]}") 
print(f"Total Adult | {adult_count} | Total Child | {child_count} | Total Senior Count {senior_count}")                 
```
