## object oriented programing

```python
class calculator:
    @staticmethod
    def add(num, num2):
        return num + num2
    
    @staticmethod
    def multiplication(a, b):
        return a * b
    
    @staticmethod
    def subtraction(a, b):
        return a - b
    
    @staticmethod
    def modulus(a, b):
        return a % b
    
    @staticmethod
    def reminder(a, b):
        return a / b

cal = calculator()
# ✅ Correct way to call them:
print(cal.multiplication(2, 4))
print(cal.subtraction(2, 4))
print(cal.modulus(3, 2))
print(cal.reminder(2, 4))
```

## polymorphism

```python
class Bird:
    def fly(self):
        print("birds are flying....")
class Airplane:
    def fly(self):
        print("airplane is flying....")
    

flying = Bird()
flying.fly()
b = Airplane()
b.fly()
                
```

### runtime polymorphism -- method overriding


```python
class Audio:
    def play(self):
        print("audio is playing...")
class Video(Audio):
    def play(self):
        print("video is playing...")        

Music = Audio()        
audio = Video()
Music.play()
audio.play()
```

## inheritance

```python
class Parent:
    def show(self):
        print("i am parent")
class Child(Parent):
    def display(self):
        print("i am child class")        
a = Child()
a.show()
a.display()        
```

## multiple inheritacne

```python
class Mother:
    def show_Mother(Self):
        print("this is mother class")
class Father:
    def show_Father(self):
        print("this is father class")
class Child(Father,Mother):
    def show_Child(Self):
        print("i am child")        

child = Child()
child.show_Father()
```

## multi level inheritance

```python
class Grandpa:
    def dada(slef):
        print("i am dada abu")
class parents(Grandpa):
    def amiabu(self):
        print("i am your maan baap")
class Child(parents):
    def myself(self):
        print("i am haroon") 

child = Child()
child.dada()
child.amiabu()
child.myself()
```
